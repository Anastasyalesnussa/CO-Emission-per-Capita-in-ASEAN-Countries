## About the Dataset

### **Description**
This dataset contains the **CO₂ emissions per capita** for ASEAN countries.
It represents carbon dioxide (CO₂) emissions resulting from the burning of fossil fuels and industrial processes. This includes emissions from transportation, electricity generation, and heating, but excludes those from land-use changes.

Data Sources: Global Carbon Budget (2024)Population based on various sources (2024) – with major processing by Our World in Data
